```python
import pandas as pd

df=pd.read_csv("../../Data/imputed.csv")
geo_df=pd.read_csv("../../Data/geo_df")
```

```python
geo_df.info()
```

```python
geo_df = geo_df.astype({"GEOID10": "string","PUMACE10":"string"})
```

```python
df = df.astype({"PUMA":"string"})
```

```python
df["GEOID10"]= "170"+df["PUMA"]
```

```python
df["PUMACE10"]= df["PUMA"]
```

```python
df
```

```python
geo_df.info()
```

```python
geo_df["GEOID10"]= geo_df["GEOID10"]+".0"
```

```python
geo_df["PUMACE10"]= geo_df["PUMACE10"]+".0"
```

```python
geo_df
```

```python
counties_df=df.merge(geo_df,how= "inner",on="PUMACE10")
```

```python
counties_df
```

```python
counties_df_2=df.merge(geo_df,how= "inner",on="GEOID10")
```

```python
counties_df_2
```

```python
Will_county=counties_df_2[counties_df_2["County_names"].str.match("Will")]
```
